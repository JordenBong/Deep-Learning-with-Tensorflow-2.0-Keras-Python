```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
import PIL
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
```

```python
dataset_url = "https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz"
data_dir = tf.keras.utils.get_file('flower_photos', origin=dataset_url, cache_dir='.', untar=True)
```

```python
# Create a dict for all flowers
flowers_image_dict = {
    'roses': list(data_dir.glob('*/roses/*')), 
    'daisy': list(data_dir.glob('*/daisy/*')), 
    'dandelion': list(data_dir.glob('*/dandelion/*')), 
    'sunflowers': list(data_dir.glob('*/sunflowers/*')), 
    'tulips': list(data_dir.glob('*/tulips/*')), 
}
```

```python
flowers_labels_dict = {
    'roses': 0, 
    'daisy': 1,
    'dandelion': 2, 
    'sunflowers': 3, 
    'tulips': 4,
}
```

```python
# Create OpenCV Object - direct convert image into 3D np-array
img = cv2.imread(str(flowers_image_dict['roses'][0])) # Return numpy arrays
img.shape
```

```python
cv2.resize(img, (180,180)).shape
```

```python
X, y = [], []

for flower_name, images in flowers_image_dict.items():
    for image in images:
        img = cv2.imread(str(image))
        resized_image = cv2.resize(img, (180,180))
        X.append(resized_image)
        y.append(flowers_labels_dict[flower_name])
```

```python
type(X), type(y)
```

```python
X = np.array(X)
y = np.array(y)

type(X), type(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y,
    random_state=0
)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
# Normalization
X_train_scaled = X_train / 255
X_test_scaled = X_test / 255
```

```python
model = Sequential([
    # Convolution
    layers.Input(shape=(180, 180, 3)),
    layers.Conv2D(16, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(32, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    
    # Classification
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(5),
])

model.compile(optimizer='adam',
             loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
             metrics=['accuracy'])

model.fit(X_train_scaled, y_train, epochs=10)
```

```python
model.evaluate(X_test_scaled, y_test)
```

**Overfitting happenning here ! How to know that?** You may notice that acc for training is 0.9983 but acc for testing only 0.6627.

```python
predictions = model.predict(X_test_scaled)
predictions
```

```python
data_augmentation = keras.Sequential([
    # Random Transformation API from keras
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
])
```

```python
model = Sequential([
    # Input Layer
    layers.Input(shape=(180, 180, 3)),
    
    # Data Augmentation
    data_augmentation,
    
    # Convolution
    layers.Conv2D(16, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(32, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),

    # Dropout Regularization
    layers.Dropout(0.2),
    
    # Classification
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(5),
])

model.compile(optimizer='adam',
             loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
             metrics=['accuracy'])

model.fit(X_train_scaled, y_train, epochs=30)
```

```python
model.evaluate(X_test, y_test)
```
